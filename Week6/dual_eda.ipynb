```python
import pandas as pd
```

```python
sample1 = pd.read_csv('data2/rerere.csv')
sample2 = pd.read_csv('data2/rerere2.csv')
sample3 = pd.read_csv('data2/rerere3.csv')
```

```python
df = pd.concat([sample1, sample2, sample3], axis=0)
```

```python
groupby = df.groupby(['사업장명', '자료생성년월'])
```

```python
first = groupby.first()
```

```python
first = first.reset_index()
first.head(10)
first.info()
```

```python
first = first.dropna()
first.reset_index(drop=True)

first.head(10)
```

```python
second = first.groupby('사업장명')['자료생성년월'].nunique()
second.head(10)
dropindex = second[second == 2].index
dropindex2 = second[second == 1].index
```

```python
third = first[first['사업장명'].isin(dropindex) == False]
third = third[third['사업장명'].isin(dropindex2) == False]
```

```python
fourth = third.groupby('사업장명')['자료생성년월'].nunique()
```

```python
fourth
first.info()
```

```python
real = first[first['사업장명'].isin(fourth.index) == True]
```

```python
real.info()
```

```python
temp_df = real.copy()

temp_df = temp_df.drop(['자료생성년월','가입자수','신규','상실','평균월급','평균연봉','고지금액','인당고지금액'], axis=1)
temp_df_groupby = real.groupby(['사업장명'])

temp_df = temp_df.groupby(['사업장명']).first()

df_201912 = real[real['자료생성년월'] == 201912]
df_202001 = real[real['자료생성년월'] == 202001]
df_202002 = real[real['자료생성년월'] == 202002]

df_201912.index = df_201912['사업장명']
df_202001.index = df_202001['사업장명']
df_202002.index = df_202002['사업장명']

temp_df["2020신규"] = temp_df_groupby['신규'].sum()
temp_df["2020상실"] = temp_df_groupby['상실'].sum()
temp_df["예상평균연봉"] = temp_df_groupby['평균연봉'].mean()
temp_df["월급201912"] = df_201912['평균월급']
temp_df["월급202001"] = df_202001['평균월급']
temp_df["월급202002"] = df_202002['평균월급']
temp_df["인원201912"] = df_201912['가입자수']
temp_df["인원202001"] = df_202001['가입자수']
temp_df["인원202002"] = df_202002['가입자수']

temp_df.head(10)
```
